```python
# ======================================
# ðŸ¥ Healthcare Data Governance Project
# Dataset: Hospital Readmissions (Kaggle)
# Author: Ksheeraja (Shiro)
# ======================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- STEP 1: Load Dataset ---
df = pd.read_csv("train.csv")
print("Shape:", df.shape)
df.head()

# --- STEP 2: Basic Summary ---
print("\n--- Dataset Overview ---")
print(df.info())

# --- STEP 3: Define Governance Checks ---
# These checks align with your dataset's structure

def check_missing_values(data):
    """Check for missing values"""
    missing = data.isnull().sum().reset_index()
    missing.columns = ['Column', 'Missing_Count']
    missing['Missing_%'] = (missing['Missing_Count'] / len(data)) * 100
    return missing[missing['Missing_Count'] > 0]

def check_negative_values(data):
    """Check for negative numeric values where not expected"""
    numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns
    negatives = {}
    for col in numeric_cols:
        negatives[col] = (data[col] < 0).sum()
    return pd.DataFrame(list(negatives.items()), columns=["Column", "Negative_Count"])

def check_zero_variance(data):
    """Check for columns with no variance (all True or all False)"""
    zero_var = data.columns[data.nunique() <= 1].tolist()
    return zero_var

def check_outliers(data, column, threshold=3):
    """Detect outliers using Z-score"""
    if column in data.columns:
        z_scores = (data[column] - data[column].mean()) / data[column].std()
        return data[np.abs(z_scores) > threshold]
    else:
        print(f"Column {column} not found.")
        return pd.DataFrame()

# --- STEP 4: Apply Quality Checks ---
missing_df = check_missing_values(df)
negatives_df = check_negative_values(df)
zero_var_cols = check_zero_variance(df)
outliers_stay = check_outliers(df, 'time_in_hospital')

# --- STEP 5: Summarise Results ---
summary = pd.DataFrame({
    'Rule': [
        'Missing Values',
        'Negative Values',
        'Zero Variance Columns',
        'Outliers in Time in Hospital'
    ],
    'Count': [
        missing_df['Missing_Count'].sum() if not missing_df.empty else 0,
        negatives_df['Negative_Count'].sum(),
        len(zero_var_cols),
        len(outliers_stay)
    ]
})

print("\n--- Data Quality Summary ---")
print(summary)

# --- STEP 6: Visualise Summary ---
plt.figure(figsize=(8,5))
plt.barh(summary['Rule'], summary['Count'], color='teal')
plt.title('Hospital Readmissions Data Quality Summary')
plt.xlabel('Number of Issues')
plt.show()

# --- STEP 7: Export for Power BI ---
summary.to_csv("data_quality_summary.csv", index=False)
missing_df.to_csv("missing_values.csv", index=False)
negatives_df.to_csv("negative_values.csv", index=False)

print("\n Files exported: data_quality_summary.csv, missing_values.csv, negatives_values.csv")

# --- STEP 8: Governance Policy Example ---
print("""
--- Example Governance Framework ---
1. All patient records must contain complete data for critical features.
2. Numeric columns (e.g., lab procedures, medications) must be non-negative.
3. Columns with zero variance should be reviewed for redundancy.
4. Time in hospital outliers (Z > 3) require manual review.
5. Data audits occur monthly and are reviewed by:
   - Data Owner: Hospital Quality Lead
   - Data Steward: Compliance Analyst
   - Data Custodian: IT Systems Admin
""")
```
